## Q2 — K-Means Clustering + Elbow Method | Iris Dataset

**Question:** Load Iris Dataset. Apply **K-Means Algorithm** using Python to group similar data points into clusters. Determine optimal number of clusters using **Elbow Method**. Visualize clustering results and analyze cluster characteristics.

**Dataset:** `sklearn.datasets.load_iris()` (built-in)

**Imports needed:** `load_iris`, `KMeans`, `silhouette_score`, `matplotlib`

```python
# STEP 1 — DATA
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

iris   = load_iris()
X      = iris.data
print('Dataset shape:', X.shape)

# STEP 2 — PREPARE (Elbow Method — plot inertia for each K)
inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertia.append(km.inertia_)

plt.figure(figsize=(8,4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method — Find Best K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (WCSS)')
plt.show()

# STEP 3 — MODEL (K=3 is elbow point for Iris)
model  = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = model.fit_predict(X)

# STEP 4 — OUTPUT
print('Silhouette Score (K=3):', round(silhouette_score(X, labels), 3))
print('Cluster sizes:', dict(zip(*np.unique(labels, return_counts=True))))

plt.figure(figsize=(7,5))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],
            color='red', marker='X', s=200, label='Centers')
plt.title('K-Means Clusters (K=3) — Sepal Length vs Width')
plt.xlabel('Sepal Length'), plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

## Q3 — K-Means Clustering + Silhouette Analysis | Iris Dataset

**Question:** Load Iris Dataset. Apply **K-Means Algorithm** using Python to group similar data points into clusters. Determine optimal number of clusters using **Silhouette Analysis**. Visualize clustering results and analyze cluster characteristics.

**Dataset:** `sklearn.datasets.load_iris()` (built-in)

**Imports needed:** `load_iris`, `KMeans`, `silhouette_score`, `matplotlib`

```python
# STEP 1 — DATA
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

iris   = load_iris()
X      = iris.data
print('Dataset shape:', X.shape)

# STEP 2 — PREPARE (Silhouette score for each K)
scores = []
for k in range(2, 11):
    km     = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.figure(figsize=(8,4))
plt.plot(range(2, 11), scores, marker='s', color='green')
plt.title('Silhouette Analysis — Best K has highest score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()

# STEP 3 — MODEL (best K = highest silhouette score)
best_k = list(range(2, 11))[scores.index(max(scores))]
print('Best K:', best_k, '| Silhouette Score:', round(max(scores), 3))
model  = KMeans(n_clusters=best_k, random_state=42, n_init=10)
labels = model.fit_predict(X)

# STEP 4 — OUTPUT
plt.figure(figsize=(7,5))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],
            color='red', marker='X', s=200, label='Centers')
plt.title(f'K-Means Clusters (K={best_k}) — Best by Silhouette')
plt.xlabel('Sepal Length'), plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

## Q4A — One Sample T-Test | Student Scores vs μ = 70

**Question:** Consider a scenario where you have test scores from a sample of students and you want to compare the mean of these scores with a hypothesized population mean.

- **Student Score = [72, 88, 64, 74, 67, 79, 85, 75, 89, 77]**
- Apply **One Sample T-Test** using Python. Assume hypothesized mean = **70**.
- Formulate Null and Alternative Hypothesis. Interpret the results and draw conclusion.

**Dataset:** Manually provided list of scores

**Imports needed:** `scipy.stats`, `numpy`

```python
# STEP 1 — DATA
from scipy import stats
import numpy as np

data = [72, 88, 64, 74, 67, 79, 85, 75, 89, 77]
mu   = 70   # Hypothesized population mean

# STEP 2 — PREPARE / HYPOTHESIS
print('H0 (Null Hypothesis)     : Mean score = 70  (no significant difference)')
print('H1 (Alt Hypothesis)      : Mean score ≠ 70  (significant difference exists)')
print(f'Sample Mean = {np.mean(data):.2f}')

# STEP 3 — ONE SAMPLE T-TEST
t, p = stats.ttest_1samp(data, mu)

# STEP 4 — OUTPUT
print(f'\nT-Statistic = {t:.4f}')
print(f'P-Value     = {p:.4f}')
print()
if p < 0.05:
    print('✅ Reject H0 — Sample mean is significantly different from 70')
    print('   → There IS a significant difference at 5% significance level.')
else:
    print('❌ Fail to Reject H0 — No significant difference from mean 70')
    print('   → The sample mean is NOT significantly different from 70.')
```

## Q5A — Chi-Square Test | Aptitude vs Job Proficiency

**Question:** The employee's aptitude and job proficiency scores are as follows:
- **aptitude = [85, 65, 50, 68, 87, 74, 65, 96, 68, 94, 73, 84, 85, 87, 91]**
- **jobprof  = [70, 90, 80, 89, 88, 86, 78, 67, 86, 90, 92, 94, 99, 93, 87]**

Perform **Chi-Square Test** to study correlation between aptitude and job proficiency. Formulate Null and Alternative Hypothesis. Interpret results and draw conclusions.

**Dataset:** Manually provided aptitude and job proficiency scores

**Imports needed:** `scipy.stats`, `pandas`, `numpy`

```python
# STEP 1 — DATA
import pandas as pd
import numpy as np
from scipy import stats

aptitude = [85, 65, 50, 68, 87, 74, 65, 96, 68, 94, 73, 84, 85, 87, 91]
jobprof  = [70, 90, 80, 89, 88, 86, 78, 67, 86, 90, 92, 94, 99, 93, 87]

# STEP 2 — PREPARE (convert continuous to categories for chi-square)
apt_cat = ['High' if x >= 75 else 'Low' for x in aptitude]
job_cat = ['High' if x >= 85 else 'Low' for x in jobprof]
df      = pd.DataFrame({'Aptitude': apt_cat, 'JobProficiency': job_cat})
table   = pd.crosstab(df['Aptitude'], df['JobProficiency'])

print('H0 (Null Hypothesis) : Aptitude and Job Proficiency are INDEPENDENT')
print('H1 (Alt Hypothesis)  : They are DEPENDENT (significantly related)')
print('\nContingency Table:')
print(table)

# STEP 3 — CHI-SQUARE TEST
chi2, p, dof, expected = stats.chi2_contingency(table)

# STEP 4 — OUTPUT
print(f'\nChi2 Statistic = {chi2:.4f}')
print(f'P-Value        = {p:.4f}')
print(f'Degrees of Freedom = {dof}')
print()
if p < 0.05:
    print('✅ Reject H0 — Aptitude and Job Proficiency ARE significantly related')
else:
    print('❌ Fail to Reject H0 — No significant relationship found')
```

## Q8A — Paired T-Test | Before vs After Remedial Lectures

**Question:** By seeing poor performance in first class test, remedial lectures were arranged. Department claims performance improved. Apply **Paired T-Test** to determine if there is a significant difference in mean test scores before and after remedial lectures.

- **Test1 (Before) = [85, 68, 67, 84, 98, 60, 94, 80, 94, 98, 95, 80, 85, 87, 75]**
- **Test2 (After)  = [70, 90, 80, 89, 88, 86, 78, 87, 90, 86, 92, 94, 99, 93, 86]**

**Dataset:** Manually provided test scores (before and after)

**Imports needed:** `scipy.stats`, `numpy`

```python
# STEP 1 — DATA
from scipy import stats
import numpy as np

test1 = [85, 68, 67, 84, 98, 60, 94, 80, 94, 98, 95, 80, 85, 87, 75]  # Before remedial
test2 = [70, 90, 80, 89, 88, 86, 78, 87, 90, 86, 92, 94, 99, 93, 86]  # After remedial

# STEP 2 — PREPARE / HYPOTHESIS
print('H0 (Null Hypothesis)     : No significant difference in scores (remedial had no effect)')
print('H1 (Alt Hypothesis)      : Significant difference exists (remedial improved performance)')
print(f'Mean Before Remedial = {np.mean(test1):.2f}')
print(f'Mean After Remedial  = {np.mean(test2):.2f}')

# STEP 3 — PAIRED T-TEST
t, p = stats.ttest_rel(test1, test2)

# STEP 4 — OUTPUT
print(f'\nT-Statistic = {t:.4f}')
print(f'P-Value     = {p:.4f}')
print()
if p < 0.05:
    print('✅ Reject H0 — Significant difference found!')
    if np.mean(test2) > np.mean(test1):
        print('   → Remedial lectures DID improve student performance.')
    else:
        print('   → Scores changed significantly, but mean decreased.')
else:
    print('❌ Fail to Reject H0 — No significant improvement after remedial.')
```

## Q10B / Q12B / Q13B — Principal Component Analysis (PCA) | Iris Dataset

**Question:** Load the Iris dataset. Perform **Principal Component Analysis (PCA)** using Python on a dataset to reduce dimensionality. Select appropriate number of principal components. Visualize the data in the reduced-dimensional space.

**Dataset:** `sklearn.datasets.load_iris()` (built-in — 4 features)

**Imports needed:** `load_iris`, `StandardScaler`, `PCA`, `matplotlib`

```python
# STEP 1 — DATA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

iris     = load_iris()
X        = iris.data
print('Original shape:', X.shape, '(150 samples × 4 features)')

# STEP 2 — PREPARE (ALWAYS scale before PCA!)
X_scaled = StandardScaler().fit_transform(X)

# STEP 3 — PCA (find optimal components)
pca_full = PCA()
pca_full.fit(X_scaled)
print('Variance per component:', pca_full.explained_variance_ratio_.round(3))
print('Cumulative variance:   ', np.cumsum(pca_full.explained_variance_ratio_).round(3))
print('→ PC1+PC2 explain', round(sum(pca_full.explained_variance_ratio_[:2])*100, 1), '% variance')

# Scree Plot
plt.figure(figsize=(8,4))
plt.bar(range(1,5), pca_full.explained_variance_ratio_, label='Individual', color='steelblue')
plt.plot(range(1,5), np.cumsum(pca_full.explained_variance_ratio_), 'ro-', label='Cumulative')
plt.axhline(y=0.95, color='green', linestyle='--', label='95% threshold')
plt.title('PCA — Scree Plot (Iris)')
plt.xlabel('Principal Component'), plt.ylabel('Explained Variance Ratio')
plt.legend(), plt.show()

# Reduce to 2D (PC1 + PC2 capture 95%+ variance)
pca_2d   = PCA(n_components=2)
X_2d     = pca_2d.fit_transform(X_scaled)

# STEP 4 — VISUALIZE
plt.figure(figsize=(7,5))
colors = ['red','blue','green']
for i, name in enumerate(iris.target_names):
    mask = iris.target == i
    plt.scatter(X_2d[mask,0], X_2d[mask,1], label=name, color=colors[i], alpha=0.7)
plt.title('Iris Dataset in 2D (PCA reduced)')
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}% var)')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}% var)')
plt.legend(), plt.show()
print('Reduced shape:', X_2d.shape, '→ 4D → 2D')
```

## Q13A — One-Way ANOVA | Class A vs B vs C Exam Scores

**Question:** Consider a dataset with exam scores of students from three different classes A, B, and C.

- **Class A = [85, 90, 88, 82, 87]**
- **Class B = [76, 78, 80, 81, 75]**
- **Class C = [92, 88, 94, 89, 90]**

Perform **One-Way ANOVA Test** using Python to determine if there is a significant difference in mean exam scores among these classes. Formulate Null and Alternative Hypothesis. Interpret results and draw conclusions.

**Dataset:** Manually provided exam scores

**Imports needed:** `scipy.stats`, `numpy`, `matplotlib`

```python
# STEP 1 — DATA
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

classA = [85, 90, 88, 82, 87]
classB = [76, 78, 80, 81, 75]
classC = [92, 88, 94, 89, 90]

# STEP 2 — PREPARE / HYPOTHESIS
print('H0 (Null Hypothesis)     : All class means are EQUAL (μA = μB = μC)')
print('H1 (Alt Hypothesis)      : At least one class mean is DIFFERENT')
print(f'Mean A = {np.mean(classA):.1f}')
print(f'Mean B = {np.mean(classB):.1f}')
print(f'Mean C = {np.mean(classC):.1f}')

# STEP 3 — ONE-WAY ANOVA TEST
f, p = stats.f_oneway(classA, classB, classC)

# STEP 4 — OUTPUT
print(f'\nF-Statistic = {f:.4f}')
print(f'P-Value     = {p:.4f}')
print()
if p < 0.05:
    print('✅ Reject H0 — Significant difference in mean scores among classes')
    print('   → At least one class has a significantly different mean score.')
else:
    print('❌ Fail to Reject H0 — No significant difference among classes')

# Visualization
plt.figure(figsize=(7,5))
plt.boxplot([classA, classB, classC], labels=['Class A','Class B','Class C'])
plt.title('One-Way ANOVA — Exam Scores by Class')
plt.ylabel('Scores')
plt.show()
```

## Q14A — Two Sample T-Test | Group 1 vs Group 2 (Task Completion Time)

**Question:** The company is assessing the difference in time to complete a task between two groups of employees. State the hypothesis and do the **Two Sample T-Test** using Python.

- **Group I (0-1 year exp) = [85, 95, 100, 80, 90, 97, 104, 95, 88, 92, 94, 99]**
- **Group II (1-2 year exp) = [83, 85, 96, 92, 100, 104, 94, 95, 88, 90, 93, 94]**

**Dataset:** Manually provided task completion times

**Imports needed:** `scipy.stats`, `numpy`, `matplotlib`

```python
# STEP 1 — DATA
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

group1 = [85, 95, 100, 80, 90, 97, 104, 95, 88, 92, 94, 99]   # 0-1 yr experience
group2 = [83, 85, 96, 92, 100, 104, 94, 95, 88, 90, 93, 94]   # 1-2 yr experience

# STEP 2 — PREPARE / HYPOTHESIS
print('H0 (Null Hypothesis) : No difference in task completion time between groups (μ1 = μ2)')
print('H1 (Alt Hypothesis)  : Significant difference exists (μ1 ≠ μ2)')
print(f'Mean Group 1 = {np.mean(group1):.2f}')
print(f'Mean Group 2 = {np.mean(group2):.2f}')

# STEP 3 — TWO SAMPLE T-TEST (Welch's — doesn't assume equal variance)
t, p = stats.ttest_ind(group1, group2, equal_var=False)

# STEP 4 — OUTPUT
print(f'\nT-Statistic = {t:.4f}')
print(f'P-Value     = {p:.4f}')
print()
if p < 0.05:
    print('✅ Reject H0 — Significant difference in task completion time')
else:
    print('❌ Fail to Reject H0 — No significant difference between groups')

# Visualization
plt.figure(figsize=(7,5))
plt.boxplot([group1, group2], labels=['Group 1 (0-1yr)','Group 2 (1-2yr)'])
plt.title('Two Sample T-Test — Task Time by Experience')
plt.ylabel('Time (minutes)')
plt.show()
```

### 📋 VLOOKUP in Excel — Implementation Steps

**Step 1 — Enter data in Excel** (columns: Supplier ID | Part Number | Part Name | Part Price | Status)

**Step 2 — VLOOKUP Syntax:** `=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])`

**Step 3 — Example formulas:**
- Part Name for A002: `=VLOOKUP("A002", B2:E21, 2, FALSE)` *(search in Part Number col, return Part Name)*

**Step 4 — For reverse lookup (finding Supplier by Part Name), use INDEX-MATCH:**

### 🐍 Python VLOOKUP Implementation

```python
# STEP 1 — DATA (full Parts dataset)
import pandas as pd

data = {
    'Supplier_ID': ['SP301','SP302','SP303','SP304','SP305','SP306','SP307','SP308',
                    'SP309','SP310','SP311','SP312','SP313','SP314','SP315','SP316',
                    'SP317','SP318','SP319','SP320'],
    'Part_Number': ['A001','A002','A003','A004','A005','A006','A007','A008',
                    'A009','A010','A011','A012','A013','A014','A015','A016',
                    'A017','A018','A019','A020'],
    'Part_Name':   ['water','altenator','air filter','wheel bearing','muffler',
                    'oil pan','brake pads','rotors','headlight','brake',
                    'Strut','Deive','CV Boot Kit','Oil Pump','Oil Filter',
                    'Fuel Filter','Tie Rod End','Ball Joint','Steering Rack','Piston'],
    'Part_Price':  [6800,3800,4500,3582,1600,1005,6500,8549,6500,1500,
                    4500,1580,2650,4660,4350,1280,1800,2500,2700,4500],
    'Status':      ['In','In','In','In stock','In','Out of stock','In','Out of stock',
                    'In','In','In','In','In stock','In','In','In',
                    'In stock','In','Out of stock','Out of stock']
}
df = pd.DataFrame(data)
print('Parts Dataset:')
print(df.to_string(index=False))

# STEP 2 — VLOOKUP helper function
def vlookup(df, search_val, search_col, return_col):
    result = df[df[search_col] == search_val][return_col]
    return result.values[0] if len(result) > 0 else 'NOT FOUND'

# STEP 3 — Q17B Queries
print('\n=== Q17B VLOOKUP Results ===')
print('1. Part Name for A002         :', vlookup(df, 'A002',       'Part_Number', 'Part_Name'))
print('2. Supplier ID for Ball Joint :', vlookup(df, 'Ball Joint', 'Part_Name',   'Supplier_ID'))
print('3. Price for muffler          :', vlookup(df, 'muffler',    'Part_Name',   'Part_Price'))
print('4. Status of A008             :', vlookup(df, 'A008',       'Part_Number', 'Status'))

# STEP 4 — Q20B Queries
print('\n=== Q20B VLOOKUP Results ===')
print('1. Part Name for A016         :', vlookup(df, 'A016',       'Part_Number', 'Part_Name'))
print('2. Supplier ID for Oil Pump   :', vlookup(df, 'Oil Pump',   'Part_Name',   'Supplier_ID'))
print('3. Price for Brake            :', vlookup(df, 'brake',      'Part_Name',   'Part_Price'))
print('4. Status of A020             :', vlookup(df, 'A020',       'Part_Number', 'Status'))
```

## Q19B — What-If Analysis | Goal Seek (Excel)

**Question:** A student's current grade is **65** (average of all semester work) and needs **at least 72** to pass. Final exam is not yet completed. Apply **What-If Analysis using Goal Seek** to determine the marks needed in Final Exam to complete the goal.

**Semester Work Grade:**

| Component         | Marks |
|-------------------|-------|
| Paper Presentation| 58    |
| Case Study        | 70    |
| Assignment        | 72    |
| Practical         | 60    |
| Final Exam        | 60    |
| **Final Grade**   | **65**|

---
### 📋 Excel Goal Seek — Implementation Steps

**Step 1 — Enter Data in Excel:**
| A (Component) | B (Marks) |
|---|---|
| Paper Presentation | 58 |
| Case Study | 70 |
| Assignment | 72 |
| Practical | 60 |
| Final Exam | 60 ← **this cell will change (input cell)** |
| Final Grade | =AVERAGE(B1:B5) ← **formula cell** |

**Step 2 — Apply Goal Seek:**
1. Click on the **Final Grade** cell (B6 = formula cell)
2. Go to `Data → What-If Analysis → Goal Seek`
3. In the Goal Seek dialog:
   - **Set cell:** B6 (Final Grade formula cell)
   - **To value:** 72 (desired final grade)
   - **By changing cell:** B5 (Final Exam cell)
4. Click **OK**

**Step 3 — Read Result:**
- Excel will calculate the value needed in Final Exam (B5) to make Final Grade (B6) = 72
- The answer will appear in B5 automatically

---
### 🐍 Python Equivalent (Goal Seek Calculation)

```python
# Python equivalent of Excel Goal Seek
import numpy as np
from scipy.optimize import brentq

# STEP 1 — DATA
grades = {
    'Paper Presentation': 58,
    'Case Study':         70,
    'Assignment':         72,
    'Practical':          60,
    'Final Exam':         60   # This is what we want to find
}
target_grade = 72

# STEP 2 — DEFINE FUNCTION
fixed_scores = [58, 70, 72, 60]   # all except Final Exam
def final_grade(final_exam_score):
    all_scores = fixed_scores + [final_exam_score]
    return np.mean(all_scores)   # AVERAGE of all 5

# STEP 3 — GOAL SEEK (find final_exam such that grade = 72)
def equation(x):
    return final_grade(x) - target_grade

required_final = brentq(equation, 0, 100)

# STEP 4 — OUTPUT
print('Current Grades:')
for k, v in grades.items():
    print(f'  {k}: {v}')
print(f'\nCurrent Final Grade = {final_grade(60):.1f}')
print(f'Target Final Grade  = {target_grade}')
print(f'\n✅ Final Exam score needed = {required_final:.2f}')
print(f'   Verify: Average with {required_final:.2f} = {final_grade(required_final):.2f}')
```

**Question:** Create CSV file from given data. Read the data from CSV files into a data frame. Perform data pre-processing tasks such as handling **missing values** and **outliers**.

**Dataset (given in question):**

| Rollno | Name   | Age | Marks | Class |
|--------|--------|-----|-------|-------|
| 1      | Sudin  | 44  | 47    | FY    |
| 2      | Shaima | 46  | 86    | SY    |
| 3      | Raina  | 27  | 45    | TY    |
| 4      | Paul   | 38  | (missing) | SY |
| 5      | Rahul  | 46  | 45    | FY    |
| 6      | Gopal  | (missing) | 67 | TY |
| 7      | Yatin  | 59  | 45    | FY    |
| 8      | Jim    | 36  | 34    | FY    |
| 9      | Nima   | 45  | 32    | TY    |

**Imports needed:** `pandas`, `numpy`

```python
# STEP 1 — DATA (Create CSV then read it)
import pandas as pd
import numpy as np

csv_content = """Rollno,Name,Age,Marks,Class
1,Sudin,44,47,FY
2,Shaima,46,86,SY
3,Raina,27,45,TY
4,Paul,38,,SY
5,Rahul,46,45,FY
6,Gopal,,67,TY
7,Yatin,59,45,FY
8,Jim,36,34,FY
9,Nima,45,32,TY"""

with open('students.csv', 'w') as f:
    f.write(csv_content)
df = pd.read_csv('students.csv')

print('--- Original Data ---')
print(df)
print('\nMissing Values:')
print(df.isnull().sum())

# STEP 2 — HANDLE MISSING VALUES
df['Marks'].fillna(df['Marks'].mean(), inplace=True)   # fill Marks with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)       # fill Age with mean
print('\n--- After Filling Missing Values ---')
print(df)

# STEP 3 — HANDLE OUTLIERS (IQR method for numeric columns)
numeric_cols = ['Age', 'Marks']
for col in numeric_cols:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    if len(outliers):
        print(f'\nOutliers in {col}: {len(outliers)} rows → removed')
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    else:
        print(f'No outliers in {col}')

# STEP 4 — OUTPUT
print(f'\n--- Final Clean Data ({len(df)} rows) ---')
print(df.reset_index(drop=True))
```

### 🐍 Python Implementation (Equivalent to Excel Pivot Table)

```python
# STEP 1 — DATA
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date':   ['03-Jan-20','14-Jan-20','21-Jan-20','30-Jan-20','07-Feb-20',
               '13-Feb-20','22-Feb-20','01-Mar-20','13-Mar-20','23-Mar-20'],
    'Color':  ['Red','Blue','Green','Blue','Green','Red','Blue','Green','Blue','Blue'],
    'Region': ['West','South','West','North','North','South','East','West','East','North'],
    'Units':  [1, 8, 2, 7, 8, 2, 5, 2, 8, 7],
    'Sales':  [110000, 96000, 26000, 84000, 25000, 60000, 35000, 87000, 69000, 54000]
}
df = pd.DataFrame(data)
print('Original Dataset:')
print(df)

# STEP 2 — PREPARE (Pivot Tables)

# Pivot 1 — Total Sales & Units
total_sales = df['Sales'].sum()
total_units = df['Units'].sum()
print(f'\nTotal Sales : ₹{total_sales:,}')
print(f'Total Units : {total_units}')

# Pivot 2 — Sales by Color
pt_color = pd.pivot_table(df, values='Sales', index='Color', aggfunc='sum')
print('\nSum of Sales by Color:')
print(pt_color)

# Pivot 3 — Region-wise Sales & Units
pt_region = pd.pivot_table(df, values=['Sales','Units'], index='Region', aggfunc='sum')
print('\nRegion-wise Sales & Units:')
print(pt_region)

# STEP 3 — VISUALIZE (PivotCharts)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Chart 1 — Sales by Color
pt_color['Sales'].plot(kind='bar', ax=axes[0], color=['blue','green','red'])
axes[0].set_title('Sum of Sales by Color')
axes[0].set_xlabel('Color'), axes[0].set_ylabel('Sales')
axes[0].tick_params(axis='x', rotation=0)

# Chart 2 — Region Sales (bar)
pt_region['Sales'].plot(kind='bar', ax=axes[1], color='steelblue')
axes[1].set_title('Region-wise Total Sales')
axes[1].set_xlabel('Region'), axes[1].set_ylabel('Sales')
axes[1].tick_params(axis='x', rotation=15)

# Chart 3 — Region Units (bar)
pt_region['Units'].plot(kind='bar', ax=axes[2], color='orange')
axes[2].set_title('Region-wise Total Units')
axes[2].set_xlabel('Region'), axes[2].set_ylabel('Units')
axes[2].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()
```

## 📌 Quick Cheat Sheet

| Statistical Test | Function | When to Use |
|-----------------|----------|-------------|
| One Sample T    | `stats.ttest_1samp(data, mu)` | 1 group vs fixed mean |
| Two Sample T    | `stats.ttest_ind(g1, g2)` | 2 independent groups |
| Paired T        | `stats.ttest_rel(before, after)` | Same group, 2 conditions |
| One-Way ANOVA   | `stats.f_oneway(a, b, c)` | 3+ groups comparison |
| Chi-Square      | `stats.chi2_contingency(crosstab)` | 2 categorical variables |

| ML Model        | Class | Key Evaluate |
|----------------|-------|--------------|
| Decision Tree   | `DecisionTreeClassifier` | `accuracy_score`, `export_text` |
| Logistic Reg    | `LogisticRegression` | `classification_report` |
| Linear Reg      | `LinearRegression` | `r2_score`, `mean_squared_error` |
| K-Means         | `KMeans` | `silhouette_score`, `inertia_` |
| PCA             | `PCA` | `explained_variance_ratio_` |

| Preprocessing  | Class | Output |
|---------------|-------|--------|
| Standardize   | `StandardScaler` | mean=0, std=1 |
| Normalize     | `MinMaxScaler` | range 0-1 |
| Encode labels | `LabelEncoder` | 0,1,2... |
| Dummify       | `pd.get_dummies()` | 0/1 columns |

**Decision Rule → p < 0.05: Reject H0 | p ≥ 0.05: Fail to Reject H0**
