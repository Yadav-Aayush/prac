Create a Simple REST Service to demonstrate CRUD operations with
“Student” database. Implement the service with fields viz. Student_Name, age,
Phone_no, Subject. Deploy the service and execute the operations

```python
student resource

package resource;

import java.util.*;
import javax.ws.rs.*;
import javax.ws.rs.core.MediaType;

@Path("students")
public class GenericResource {

    static Map<Integer, student> studentDB = new HashMap<>();

    // CREATE
    @POST
    @Consumes(MediaType.APPLICATION_JSON)
    @Produces(MediaType.APPLICATION_JSON)
    public student createStudent(student s) {
        studentDB.put(s.getId(), s);
        return s;
    }

    // READ ALL
    @GET
    @Produces(MediaType.APPLICATION_JSON)
    public Collection<student> getAllStudents() {
        return studentDB.values();
    }

    // READ BY ID
    @GET
    @Path("/{id}")
    @Produces(MediaType.APPLICATION_JSON)
    public student getStudent(@PathParam("id") int id) {
        return studentDB.get(id);
    }

    // UPDATE
    @PUT
    @Path("/{id}")
    @Consumes(MediaType.APPLICATION_JSON)
    @Produces(MediaType.APPLICATION_JSON)
    public student updateStudent(@PathParam("id") int id, student s) {
        studentDB.put(id, s);
        return s;
    }

    // DELETE
    @DELETE
    @Path("/{id}")
    public String deleteStudent(@PathParam("id") int id) {
        studentDB.remove(id);
        return "Student Deleted Successfully";
    }
}
```

Write a Program to implement web Service to Calculate factorial of a given
number.
Deploy the above Web service properly and call it by designing appropriate
interface in JAVA / .NET

```python
Factorial
webservice > Factorial.java
package server;

import javax.jws.WebService;
import javax.jws.WebMethod;
import javax.jws.WebParam;

@WebService(serviceName = "Factorial")
public class Factorial {
    @WebMethod(operationName = "Factorial")
    public int Factorial(@WebParam(name = "num") int num) {
        int fact = 1;
        for(int i = 1; i <= num; i++)
        {
            fact = fact * i;
        }
        return fact;
    }
}
```

Write a Program to implement web Service to identify whether number is prime
or not.
Deploy the above Web service properly and call it by designing appropriate
interface in JAVA / .NET

```python
Prime Number
webservice > PrimeService.java
package server;

import javax.jws.WebService;
import javax.jws.WebMethod;
import javax.jws.WebParam;

@WebService(serviceName = "PrimeService")
public class PrimeService {

    @WebMethod(operationName = "checkPrime")
    public String checkPrime(@WebParam(name = "num") int num) {
        int flag = 0;
        if(num <= 1)
        {
            return "Not Prime";
        }
        for(int i = 2; i <= num/2; i++)
        {
            if(num % i == 0)
            {
                flag = 1;
                break;
            }
        }
        if(flag == 0)
            return "Prime Number";
        else
            return "Not Prime Number";
    }
}

```

```python
Step 1: Update Ubuntu System
Open terminal and run:
sudo apt update && sudo apt upgrade -y
sudo apt dist-upgrade -y
sudo reboot

Step 2: Create OpenStack User
Create a new user named stack:
sudo adduser stack
Add the user to sudo group:
sudo usermod -aG sudo stack
Give sudo permission:
echo "stack ALL=(ALL) NOPASSWD: ALL" | sudo tee /etc/sudoers.d/stack
Switch to stack user:
su - stack

Step 3: Install Required Packages
Install Git and network tools:
sudo apt install git -y
sudo apt install net-tools -y

Step 4: Download OpenStack DevStack
Clone the DevStack repository:
git clone https://opendev.org/openstack/devstack
Move into the directory:
cd devstack

Step 5: Check Network Configuration
Check your IP address:
ifconfig
Note the IP address (example: 192.168.152.140).

Step 6: Create DevStack Configuration File
Create configuration file:
sudo nano local.conf
Add this content:
[[local|localrc]]
ADMIN_PASSWORD=stack
DATABASE_PASSWORD=stack
RABBIT_PASSWORD=stack
SERVICE_PASSWORD=stack
HOST_IP=192.168.152.140
LOGFILE=/opt/stack/logs/stack.sh.log
VERBOSE=True
Save and exit.

Step 7: Install OpenStack
Run the installation script:
./stack.sh
Installation takes 10–20 minutes.
When finished, the terminal shows:
Horizon is now available at http://<IP>/dashboard

Step 8: Open OpenStack Dashboard
Open browser and go to:
http://<your-ip>/dashboard
Example:
http://192.168.152.140/dashboard

Step 9: Login to OpenStack
Use credentials:
Username: admin
Password: stack

Step 10: Open OpenStack Dashboard
After login, the OpenStack Horizon dashboard appears.
You can manage:
Instances
Networks
Volumes
Images
Security Groups
```
